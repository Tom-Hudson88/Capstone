## downloading geojson file that will show +49k public charging locations throughout the US

```python
alt_fuel_sites = gpd.read_file('../../data/Source of Data - afdc.energy.gov/alt_fuel_stations.geojson')
```

### dataset includes geometry point; need to unpack lat and lon from geometry in order to utilize folium mapping

```python
alt_fuel_sites['lon'] = alt_fuel_sites.geometry.x
alt_fuel_sites['lat'] = alt_fuel_sites.geometry.y
```

```python
alt_fuel_sites.head()
```

```python
alt_fuel_sites.loc[(alt_fuel_sites['lon'] == 0)]
```

### two datapoints show US based charging station on the prime meridian; obvious keying error; removing from dataframe

```python
alt_fuel_sites.drop([43115,43123], axis=0, inplace = True)
```

### establishing 'area_center' as center of Davidson County, TN as a simple reference for Folium

```python
area_center = (36.17000, -86.78000)
```

```python

map_alt_fuel_sites = folium.Map(location =  area_center, zoom_start = 3)



#create a list of locations and pass them to FastMarkerCluster()
locations = alt_fuel_sites[['lat', 'lon']].values.tolist()
map_alt_fuel_sites.add_child(
    FastMarkerCluster(locations))



#display our map
map_alt_fuel_sites
```

```python
map_alt_fuel_sites.save('../maps/map_alt_fuel_sites.html')
```
