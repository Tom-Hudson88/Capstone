```python
renew_elec = renew_elec_gen.copy()
```

```python
renew_elec.columns
```

```python
renew_elec.rename(columns = {'Utility Scale Wind':'wind', 'Utility Scale Solar\nPhotovoltaic':
                             'solar_photovoltaic', 'Utility Scale Solar\nThermal':'solar_thermal',
                            'Utility Scale Conventional\nHydroelectric':'hydroelectric'}, inplace = True)
```

```python
renew_elec
```

```python
renew_elec.rename(columns = {'Period':'year','Utility Scale Wood and\nWood-Derived\nFuels':'wood', 'Utility Scale Landfill\nGas':
                             'methane', 'Utility Scale Biogenic\nMunicipal\nSolid Waste':'solid_waste',
                            'Utility Scale Other Waste Biomass':'other_biomass','Utility Scale Geothermal':'geothermal'}, inplace = True)
renew_elec
```

```python
renew_elec.info()
```

```python
renew_elec['wind'] = renew_elec['wind'].str.replace(',','')
renew_elec['solar_photovoltaic'] = renew_elec['solar_photovoltaic'].str.replace(',','')
renew_elec['solar_thermal'] = renew_elec['solar_thermal'].str.replace(',','')
renew_elec['wood'] = renew_elec['wood'].str.replace(',','')
renew_elec['methane'] = renew_elec['methane'].str.replace(',','')
renew_elec['solid_waste'] = renew_elec['solid_waste'].str.replace(',','')
renew_elec['other_biomass'] = renew_elec['other_biomass'].str.replace(',','')
renew_elec['hydroelectric'] = renew_elec['hydroelectric'].str.replace(',','')
renew_elec['geothermal'] = renew_elec['geothermal'].str.replace(',','')

renew_elec
```

```python
renew_elec['wind'] = renew_elec['wind'].astype(float)
renew_elec['solar_photovoltaic'] = renew_elec['solar_photovoltaic'].astype(float)
renew_elec['solar_thermal'] = renew_elec['solar_thermal'].astype(float)
renew_elec['wood'] = renew_elec['wood'].astype(float)
renew_elec['methane'] = renew_elec['methane'].astype(float)
renew_elec['solid_waste'] = renew_elec['solid_waste'].astype(float)
renew_elec['other_biomass'] = renew_elec['other_biomass'].astype(float)
renew_elec['geothermal'] = renew_elec['geothermal'].astype(float)
renew_elec['hydroelectric'] = renew_elec['hydroelectric'].astype(float)

renew_elec
```

```python
renew_elec['bio-waste'] = renew_elec['wood']+renew_elec['methane']+renew_elec['solid_waste']+renew_elec['other_biomass'].astype(float)

renew_elec
```

```python
renew_elec.info()
```

```python
cols = [4,5,6,7,10,11,12,13]
renew_elec.drop(renew_elec.columns[cols], axis=1, inplace=True)
renew_elec
```

```python
renew_elec = renew_elec.melt(id_vars='year',var_name='source_of_generation', value_name='units_generated')
```

```python
renew_elec.info()
```

```python
renew_elec['units_generated'] = renew_elec['units_generated'].astype(float)
renew_elec.head(2)
```

```python
plt.figure(figsize = (15,5)) 
ax = sns.lineplot(x = 'year', y = 'units_generated', data = renew_elec, hue = 'source_of_generation', linewidth = 3.0,
                  palette=['blue','#ffcc33','#00ff00','#ff0000','#ed872d','#a52a2a'])
 

ax.set(ylabel='Thousand Megawatt Hours', title='Electricity Generation - Renewable Sources')
plt.show();
```
