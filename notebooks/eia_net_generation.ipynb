## Clean data by dropping final 2 digits from YYYYMM column and renaming column "year"

```python
eia_net_generation['YYYYMM'] = eia_net_generation['YYYYMM'].apply(str)
```

```python
eia_net_generation['YYYYMM'] = eia_net_generation['YYYYMM'].str.replace('\d\d$','')
```

```python
eia_net_generation = eia_net_generation.rename(columns = {'YYYYMM': 'year'})
eia_net_generation.head(2)
```
