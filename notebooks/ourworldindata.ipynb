## The continents of Asia and South America do not exist as a dataset; I am selecting the countries available for a South America df and the following countries for Asia based on industrialization: China, India, Taiwan, Japan, South Korea, Indonesia, Malaysia, Singapore and Vietnam

```python
south_america = continents.loc[continents.Entity.isin(['Chile', 'Argentina', 'Brazil', 'Peru', 'Equador','Columbia','Venezuela'])]
south_america
```

```python
south_america = south_america.groupby(['Year'])['Renewables (% sub energy)'].mean()
```

```python
south_america = south_america.to_frame()
```

```python
south_america.head()
```

```python
south_america.shape
```

```python
south_america.insert(loc=0, column='Entity', value='South America')
```

```python
south_america.head()
```

```python
south_america.reset_index(inplace=True)
```

```python
south_america.head()
```

```python
south_america.info()
```

#### Note: dataset only had data for 7 countries in South America
#### Note: the column 'Renewables (% sub energy) is a mean of these countries' individual values with no weighting based on each Countries electrical usage or population

```python
asia = continents.loc[continents.Entity.isin(['China', 'India', 'Taiwan', 'South Korea', 'Japan','Indonesia','Malaysia','Singapore','Vietnam'])]
asia
```

```python
asia = asia.groupby(['Year'])['Renewables (% sub energy)'].mean()
```

```python
asia = asia.to_frame()
```

```python
asia.shape
asia.head(5)
```

```python
asia.insert(loc=0, column='Entity', value='Asia')
```

```python
asia.head()
```

```python
asia.reset_index(inplace=True)
```

```python
asia.head()
```

```python
asia.info()
```

#### Note: dataset did not have a separate 'Asia' category; I selected China, India, South Korea, Taiwan, Japan, Singapore, Indonesia, Malaysia and Vietnam as 'industrialized' countries.
#### Note: the column 'Renewables (% sub energy) is a mean of these countries' individual values with no weighting based on each Countries electrical usage or population

```python
continents = pd.concat([iceland,united_states,north_america, south_america, asia, europe, africa, australia])
```

```python
continents.reset_index(drop=True)
```

```python
plt.figure(figsize = (20,12)) 
sns.color_palette('Set2')
ax=sns.lineplot(x = 'Year', y = 'Renewables (% sub energy)', data = continents, hue = 'Entity', linewidth = 3.0)
ax.set(ylabel= 'Renewables as Percentage', title='Renewable Energy as Percentage of Total Electricity')
plt.show()
```
